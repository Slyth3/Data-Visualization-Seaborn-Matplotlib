```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y= 0.5*x +2
```

## Quick plotting 

```python
plt.plot(x,y)
plt.xlabel("bottom x")
plt.ylabel("side y")
plt.title("Hello")
plt.show()
```

### quick subplot 

```python
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(1/y,x)
```

# Best method 

```python
fig = plt.figure ()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
```

```python
fig = plt.figure ()

ax = fig.add_axes([0,0,1,1])


ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.set_title("title")

ax.plot(x,y)
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1,0.6,0.3,0.3])

ax1.set_title("I am big")
ax2.set_title("I am smol")

ax1.plot(x,y)
ax2.plot(1/x,y)
```

# Object orientated Subplots 

```python
fig, axes = plt.subplots(1,2)

axes[0].plot(x,y)
axes[1].plot(1/x,y)
```

## colour 

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
#alpha is transparency
ax.plot(x,y+4, color = "green", lw =20 , alpha = 0.5, ls = ":", marker = "s", markersize = 10) 
ax.plot(x,y+2, color = "r", lw= 0.5, marker = "o")
ax.plot(x,y, color = "#0033a1", linewidth= 4 , ls = "--")
```

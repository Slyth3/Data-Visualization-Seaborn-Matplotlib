```python
import matplotlib.pyplot as plt 
import matplotlib.ticker as ticker
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(".\\f500.csv")
```

```python
df.head()
```

```python
# plot all lines 
plt.plot()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1 ])

ax.plot( df["company"].head(), df["profits"].head() , lw = 3, ls= "--")
ax.plot( df["company"].tail(), df["profits"].tail(), marker= "o" )

ax.set_ylim(0,15000)

ax.set_xticklabels(df["company"], rotation = 90)

# wraping x labels and rotaing them 
ax.set_xticklabels(df["company"].str.replace(" ","\n"), rotation =45)
#ax.xaxis.set_major_locator(ticker.AutoLocator())

plt.show()
```
